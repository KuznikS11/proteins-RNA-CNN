```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('statistics_all.csv')[['Aminoacid', 'x size', 'y size', 'z size']]
for name, data in df.groupby('Aminoacid'):
    x_sizes = data['x size'].values
    plt.hist(x_sizes, bins=100)
    plt.xticks(range(0, 16, 1))
    plt.xlabel('Bounding box x edge length')
    plt.ylabel('Frequency')
    plt.title(f'Amino acid: {name}')
    plt.show()

    print(f'Amino acid: {name}, mean of x sizes: {np.mean(x_sizes)}')

    y_sizes = data['y size'].values
    plt.hist(y_sizes, bins=100)
    plt.xticks(range(0, 16, 1))
    plt.xlabel('Bounding box y edge length')
    plt.ylabel('Frequency')
    plt.title(f'Amino acid: {name}')
    plt.show()

    print(f'Amino acid: {name}, mean of y sizes: {np.mean(y_sizes)}')

    z_sizes = data['z size'].values
    plt.hist(z_sizes, bins=100)
    plt.xticks(range(0, 16, 1))
    plt.xlabel('Bounding box z edge length')
    plt.ylabel('Frequency')
    plt.title(f'Amino acid: {name}')
    plt.show()

    print(f'Amino acid: {name}, mean of z sizes: {np.mean(z_sizes)}')
```
