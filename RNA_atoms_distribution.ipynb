```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 10]

structure_ids = []   
for line in open('./structures lists/stucture ids homo sapiens.txt', 'r'):
    line = line.strip('\n').lower()
    structure_ids.append(line)
for line in open('./structures lists/stucture ids synthetic construct.txt', 'r'):
    line = line.strip('\n').lower()
    structure_ids.append(line)
for line in open('./structures lists/stucture ids virus.txt', 'r'):
    line = line.strip('\n').lower()
    structure_ids.append(line)

structure_ids.remove('1a9n')
structure_ids.remove('2adc')
print(len(structure_ids))



# sumed_rna = []
# num_aa = 0
# for i, structure_id in enumerate(structure_ids):
#     rna = np.load('../data/voxelized data 14x14x17 2/' + structure_id + '_rna_3D.npy')
#     num_aa+=len(rna)
    
#     sumed_rna.extend(list(map(lambda aa: [[[np.sum(aa[:7, :7, i]), np.sum(aa[:7, 7:, i])], [np.sum(aa[7:, :7, i]), np.sum(aa[7:, 7:, i])]] for i in range(17)], rna)))
    
# sumed_rna = np.array(sumed_rna)
# np.save('sumed_rna_stats', sumed_rna, allow_pickle=False)

# print(sumed_rna.shape)
# print(num_aa)

sumed_rna = np.load('sumed_rna_stats.npy')
print(sumed_rna.shape)

for l in range(17):
    f, ax = plt.subplots(2,2)
    f.suptitle(f'level:{l+1}')
    
    first_quater = list(filter(lambda x: x > 0, sumed_rna[:,l, 0,0]))
    print(f'Level {l+1}, first  quater: Min: {np.min(first_quater)}, Max: {np.max(first_quater)}, Mean: {np.mean(first_quater)}')
    ax[0,0].hist(first_quater, bins=100)
    ax[0,0].set_xlim(left=0,right=150)
    
    second_quater = list(filter(lambda x: x > 0, sumed_rna[:,l, 1,0]))
    print(f'Level {l+1}, second  quater: Min: {np.min(second_quater)}, Max: {np.max(second_quater)}, Mean: {np.mean(second_quater)}')
    ax[0,1].hist(second_quater, bins=100)
    ax[0,1].set_xlim(left=0,right=150)
    
    third_quater = list(filter(lambda x: x > 0, sumed_rna[:,l, 0,1]))
    print(f'Level {l+1}, third  quater: Min: {np.min(third_quater)}, Max: {np.max(third_quater)}, Mean: {np.mean(third_quater)}')
    ax[1,0].hist(third_quater, bins=100)
    ax[1,0].set_xlim(left=0,right=150)
    
    fourth_quater = list(filter(lambda x: x > 0, sumed_rna[:,l, 1,1]))
    print(f'Level {l+1}, fourth  quater: Min: {np.min(fourth_quater)}, Max: {np.max(fourth_quater)}, Mean: {np.mean(fourth_quater)}')
    ax[1,1].hist(fourth_quater, bins=100)
    ax[1,1].set_xlim(left=0,right=150)
    
    plt.show()
    
for l in range(17):
    floor = list(filter(lambda x: x > 0, sumed_rna[:,l,0,0] + sumed_rna[:,l,0,1] + sumed_rna[:,l,1,0] + sumed_rna[:,l,1,1]))
    print(f'Level {l+1}: Min: {np.min(floor)}, Max: {np.max(floor)}, Mean: {np.mean(floor)}')
    plt.hist(floor, bins=200)
    plt.xlim(left=0,right=200)
    plt.title(f'level:{l+1}')
    plt.show()

one = list(filter(lambda x: x > 0, map(lambda x: np.sum(x), sumed_rna)))
print(f'All: Min: {np.min(one)}, Max: {np.max(one)}, Mean: {np.mean(one)}')
plt.hist(one, bins=1000)
plt.xlim(left=0, right=200)
plt.title('All')
                
plt.show()
```
